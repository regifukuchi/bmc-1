```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

# Constantes

```python
tt = 10 #segundos
g= -9.8 #gravidade
v0 = 30 #m/s
teta = 30*math.pi/180 #graus
x0 = 0
y0 = 0
```

# Fórmulas
    Horizontal: x = x0 + v0*cos(teta)*t
    
    Vertical:   y = y0 + v0*sen(teta)*t + g/2*t²
    
    Ao isolar t em x e substituir em y, temos
    
    Vertical:   y = y0 + v0*tan(teta)v0*(x-x0) + g/2*((x-x0)/v0*cos(teta))^2

```python
t = np.linspace(start=0, stop=tt, num=100) #vetor tempo
x =  x0 + v0*math.cos(teta)*t
y =  y0 + v0*math.sin(teta)*t + g/2*t**2  
y1 = y0 + v0*math.tan(teta)*(x-x0)/v0 + g/2*((x-x0)/(v0*math.cos(teta)))**2
```

# Plot 1
Calculo da posição baseado no tempo

```python
plt.plot(x,y)
plt.show
```

# Plot 2
Calculo da posição vertical baseado na posição horizontal

```python
plt.plot(x,y1)
plt.show
```

# Plot 3
Calculo da posição vertical e horizontal no tempo

```python
fig, axs = plt.subplots(nrows=1, ncols= 2,sharex = True, figsize = (10,5), squeeze =True)
axs[0].plot(t,y,label = 'vertical')
axs[1].plot(t,x,label = 'horizontal')
plt.show()
```

# Plot 4
Estipula um mínimo

```python
#cria mínimo de zero para o eixo y, partindo do pressuposta que a particula não passará abaixo do y0
for i in range(0,10*tt):
    if y[i] < 0:
        y[i] = 0
```

```python
plt.plot(x,y)
plt.show
```
