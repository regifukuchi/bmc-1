## Exercise

A football is kicked with an angle of 30 degrees with the ground. The mass of the ball is 0.43 kg. Consider the gravitational acceleration as 9.81 m/s$^2$. The initial speed of the ball is 30 m/s and the initial height is 0 m.

Write a Jupyter notebook to find the trajectory of a ball considering the air drag proportional to the square root of the ball velocity.

$\vec{F_b} = -\frac{b}{\sqrt{v}}\vec{v}$, where $v = \sqrt{v_x^2+v_y^2}$ is the magintude of the velocity vector.

<figure><img src="../../../images/solP12018Q2_D.png\" width=200 />

## Vertical

$\frac{dv_y}{dt} = -g -\frac{b}{m\sqrt{v}}v_y$

## Horizontal

$\frac{dv_x}{dt} = -\frac{b}{m\sqrt{v}}v_x$

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Constantes

```python
g = 9.81
v0 = 30
m = 0.43
teta = 30*np.pi/180
v0y = v0*np.sin(teta) 
v0x = v0*np.cos(teta)
dt = 0.01
tt = 3.5
b = 0.006
y0 = 0
x0 = 0
```

```python
t = np.linspace(start=0, stop=tt, num=np.int(tt/dt)) #vetor tempo

ry = np.empty_like(t)
rx = np.empty_like(t)
rvy = np.empty_like(t)
rvx = np.empty_like(t)
rv = np.empty_like(t)

ry[0] = y0
rx[0] = x0
rvy[0] = v0y
rvx[0] = v0x
rv[0] = v0

for i in range(1,len(t)):
    dvxdt = -b*rvx[i-1]/(m*np.sqrt(rv[i-1]))
    rvx[i] = rvx[i-1] + dt*dvxdt
    
    dxdt = rvx[i-1]
    rx[i] = rx[i-1] + dt*dxdt
    
    
    dvydt = -g -b*rvy[i-1]/(m*np.sqrt(rv[i-1]))
    rvy[i] = rvy[i-1] + dt*dvydt
    
    dydt = rvy[i-1]          
    ry[i] = ry[i-1] + dt*dydt
    
    rv[i] = np.sqrt(rvx[i]**2 + rvy[i]**2)
```

```python
plt.figure()
plt.plot(rx,ry)
plt.ylim(0,13)
plt.xlim(0,85)
plt.show()
```
