Plot the trajectory of a particle during 10 seconds with the gravity force acting on it (gravity acceleration g = -9.81 m/s^2). Consider that the initial velocity of the particle is v0 = 30 m/s and the initial angle with the horizontal plane is 30 degrees.

```python
import numpy as np
import pandas as pd
import pylab as pl
import matplotlib.pyplot as plt
import math

%matplotlib notebook
```

```python
t = np.arange(11)
g = -9.81  # m/s^2
v0 = 30 # m/s
alfa = np.pi/6 #rad
```

## Trajectory

```python
# consider x0 and y0 = 0m
x = v0*math.cos(alfa)*t
y = v0*math.sin(alfa)*t + (g*np.multiply(t,t))/2
```

## Plot 

```python
fig, ax = plt.subplots(1, 1, figsize=(10,5), sharex=True)

ax.plot(x,y,label='trajectory')
plt.xlabel('x [m]')
plt.ylabel('y [m]')
ax.legend()
```
