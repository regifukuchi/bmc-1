Write a Jupyter notebook to compute and plot the acceleration of the particle with displacement data contained in the file pezzack.txt. Additionally, compare the computed acceleration with the true acceleration (contained in the same data).

```python
import numpy as np
import pandas as pd
import pylab as pl
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
filePath = 'F:/Desiree/Disciplinas/ControleMotor/Pezzack.txt'

data = pd.read_csv(filePath,header=3, sep='\t')
#data
```

```python
data.Time[1]
```

## Acceleration

```python
dt = data.Time[1]
rawA = np.diff(np.diff(data.Raw)/dt)/dt
noisyA = np.diff(np.diff(data.Noisy)/dt)/dt
#rawA
```

## Plot 

```python
fig, ax = plt.subplots(1, 1, figsize=(10,5), sharex=True)

ax.plot(data.Accel,c='k',label='true')
ax.plot(rawA,c='b',label='raw')
ax.plot(noisyA,c='r',label='noisy')
ax.legend()
```
