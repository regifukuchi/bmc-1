```python
import numpy as np
import matplotlib.pyplot as plt
import math

%matplotlib notebook
```

```python
dt = 0.01 #s
g = 9.81  # m/s^2
x0 = 0 #m
y0 = 0 #m
alfa = np.pi/6
v0 = 30 # m/s
v0x = v0*np.cos(alfa) # m/s
v0y = v0*np.sin(alfa) # m/s

b = 0.06
m = 0.43
```

## Trajectory

```python
t = np.arange(0,3.4,dt)
rx = np.empty_like(t)
ry = np.empty_like(t)
```

```python
vx = v0x
vy = v0y
x = x0
y = y0
```

```python
for i in range (len(t)):
    mod = np.sqrt(vx**2 + vy**2)
    
    
    #State eq
    dxdt = vx
    dvxdt = -b/(mod*m)*vx
    
    dydt = vy
    dvydt = -b/(mod*m)*vy -g
    
    #Euler
    x = x + dt*dxdt
    vx = vx + dt*dvxdt
    
    y = y + dt*dydt
    vy = vy + dt*dvydt
    
    rx[i] = x
    ry[i] = y
```

## Plot 

```python
fig, ax = plt.subplots(1, 1, figsize=(10,5), sharex=True)

ax.plot(rx,ry,label='trajectory')
plt.xlabel('x [m]')
plt.ylabel('y [m]')
ax.legend()
```
