```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
data = np.loadtxt('C:\\Users\\Fundacao\\Desktop\\Pezzack.txt', skiprows=1)
```

```python
#Atribuindo valores ás variáveis
# t = Tempo
# x = Posição
# a = Aceleração

t = data[:,0]
x = data[:,1:2]
a = data[:,3]
```

```python
print("Tempo: \n", t)
```

```python
print("Posição: \n", x)
```

```python
print("Aceleração: \n", a)
```

```python
#Primeira derivada em relação eo tempo
dt = t[1]-t[0]
v = np.diff(x, n=1, axis=0)/dt
```

```python
print ("velocidade: \n", v)
```

```python
a2 = np.diff(v, n=1, axis=0)/dt
```

```python
print ("Aceleração com ruído: \n", a2)
```

```python
#para ver a diferença entre as duas velocidades
adif = a2-a
```

```python
print ("A diferença entre a aceleração original e a com ruído é: \n", adif)
```

```python
#Para deixar ambas as variáveis com o mesmo tamanho
size_a2 = np.size(a2)
size_a2
```

```python
new_t = t[0:size_a2]
np.size(new_t)
```

```python
plt.plot(new_t, a2)
plt.show()
```

```python
plt.plot(t, a)
plt.show()
```
