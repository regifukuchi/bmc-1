```python
import numpy as np
import matplotlib.pyplot as plt
import math

%matplotlib notebook
```

```python
#Valores iniciais
dt = 0.001
t = np.arange(0,300,dt)
f = np.empty (t.shape)

b = 500
ks = 100
kp = 400
l0 = 0.01
l = 0.01

f[0] = 0
```

```python
# Para o Modelo Kelvin

x1 = f[0] - (kp+ks)*l

for i in range (1,len(t)):
    if t[i]>=60:
        l = 7e-3
     
    dx1dt = -(ks/b)*x1 -((ks**2)/b)*l -((ks*kp)/b)*l0
    x1 = x1 + dt*dx1dt
    f[i] = x1 + (kp+ks)*l
```

```python
plt.plot(t,f)
plt.show()
```
