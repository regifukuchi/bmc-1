```python

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math
```

```python
#Inserção de valores fornecidos

v0 = 30
g = -9.81
theta = np.pi/6 #Em radianos
t = np.arange(11) #0 a 10 segundos
```

```python
#Trajetória ao longo do eixo X

x = v0*math.cos(theta)*t
```

```python
#Trajetória ao longo do Eixo Y

y = v0*math.sin(theta)*t + 0.5*(g*(t*t))
```

```python
plt.plot(x,y)
plt.show()
```
