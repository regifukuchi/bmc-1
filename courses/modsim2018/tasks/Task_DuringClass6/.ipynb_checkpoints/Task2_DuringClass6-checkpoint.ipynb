## Trajetória com Força de Arrasto

```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
dt = 0.01
g = 9.81
b = 0.006
m = 0.43
```

```python
# Initial conditions
x0 = 0
y0 = 0
v0 = 30

x = x0
y = y0

ang = 30 * np.pi/180

v0 = 30
v0x = v0 * np.cos(ang)
v0y = v0 * np.sin(ang)

vx = v0x
vy = v0y
```

```python
t = np.arange(0,3.4,dt)

rx = np.empty_like(t)
ry = np.empty_like(t)


for i in range(len(t)):
    moduloV = np.sqrt(vx**2 + vy**2)
    
    # State Equation
    dxdt = vx
    dvxdt = (-b/(moduloV * m)) * vx

    dydt = vy
    dvydt = (-b/(moduloV * m)) * vy - g
    
    
    # Euler step for next item of vector
    x = x + dt * dxdt
    vx = vx + dt *dvxdt

    y = y + dt * dydt
    vy = vy + dt *dvydt
    
    rx[i] = x
    ry[i] = y
```

```python
fig, ax=plt.subplots(1, 1)

ax.plot(rx,ry,label='Trajetória')
ax.set_ylabel('y [m]')
ax.set_xlabel('x [m]')
ax.legend()
plt.tight_layout()
plt.show()
```
