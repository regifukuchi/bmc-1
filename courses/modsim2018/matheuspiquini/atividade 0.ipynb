```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time, disp, disp2, accel = np.loadtxt('values.txt', unpack=True)
```

```python
%matplotlib inline
plt.plot(time, disp)
plt.show()
```

```python
plt.plot(time,disp2)
plt.show()
```

```python
plt.plot(time, accel)
plt.show()
```

```python
dt = np.mean(np.diff(time))
v=np.diff(disp)/dt
a=np.diff(v)/dt
time2 = time[:-2]

plt.plot(time2,a)
plt.show()
```

```python
v2=np.diff(disp2)/dt
a2=np.diff(v2)/dt

plt.plot(time2,a2)
plt.show()
```
