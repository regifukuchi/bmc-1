```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x0=0;
y0=0;
vx0=30*np.cos(30*np.pi/180)
vy0=30*np.sin(30*np.pi/180)
g=9.81
dt=0.01
b=0.006
m=0.43
b=0
```

```python
t=np.arange(0,3,dt)
rx=np.empty_like(t)
ry=np.empty_like(t)
x=x0
y=y0
vx=vx0
vy=vy0

for i in range(len(t)):
    dxdt=vx
    dydt=vy
    dvxdt=-b/(m*vx)
    dvydt=-g-b/(m*vy)
    x= x+ dt*dxdt
    y= y+ dt*dydt
    vx= vx+ dt*dvxdt
    vy= vy +dt*dvydt
    rx[i]=x
    ry[i]=y
```

```python
plt.plot(rx,ry)
plt.show()
```
