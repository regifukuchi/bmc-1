Fx = -b/np.sqrt(np.sqrt(vx**2+vy**2)) * vx

Para x
x'' = (-b/np.sqrt(np.sqrt(vx**2+vy**2)) * vx)/m
vx' = (-b/np.sqrt(np.sqrt(vx**2+vy**2)) * vx)/m
x' = vx

Para y
y'' = -b/np.sqrt(np.sqrt(vx**2+vy**2)) * vy - m*g
vy' = (-b/np.sqrt(np.sqrt(vx**2+vy**2)) * vy)/m - g
y' = vy

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib notebook

x0 = 0
y0 = 0
vx0 = 30*np.cos(30*np.pi/180)
vy0 = 30*np.sin(30*np.pi/180)

x = x0
y = y0
vx = vx0
vy = vy0

g = 9.81
b = 0.006
m = 0.43
dt = 0.01


t = np.arange (0, 3, dt)

rx = np.empty_like(t)
ry = np.empty_like(t)

for i in range(len(t)): 

    dxdt = vx
    dydt = vy
    dvxdt = (-b/np.sqrt(np.sqrt(vx**2+vy**2)) * vx)/m
    dvydt = (-b/np.sqrt(np.sqrt(vx**2+vy**2)) * vy)/m - g

    x = x + dt*dxdt
    y = y + dt*dydt
    vx = vx + dt*dvxdt
    vy = vy + dt*dvydt

    rx[i] = x
    ry[i] = y
    
plt.plot(rx,ry)
plt.show()
```
